```python
#Subtracts background and displays as a black area
%run show_image.ipynb
import cv2
import numpy as np


cap = cv2.VideoCapture('Wacci_Video.mp4')
scaling_factor = 1

#Use history number of frames, use 100 previous frames in this case. higher number may lead to slower learning rate
bg_subtractor = cv2.createBackgroundSubtractorMOG2()
history = 100

learning_rate = 1.0 / history

#Continuously obtain frame
try:
    while True:
        frame = get_frame(cap, scaling_factor)
        if frame is None:
            print('Finished')
            break
        #Compares frame to previous frame, shows differences
        mask = bg_subtractor.apply(frame, learningRate = learning_rate)
        mask = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)
        img2 = cv2.hconcat([frame, frame & mask])
        show_image(img2)

except KeyboardInterrupt:
    print('Interrupted')
    
cap.release()
```
